```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Here i am reading the files from my document directory
filepath=r'C:\Users\user\Documents\correlation.csv';
file=pd.read_csv (filepath);
```

```python
file.info()
```

```python
#I am applying the pearson correlation concept 
corr_matrix = file.corr().abs()

print(corr_matrix)
# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find index of feature columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]
#Here we print the correlation matrix which is in upper-trianguar form
```

```python
#here we can see in the below output that since the correlation between the elements next to each other is less that 0.85 
#none of the elements are dropped

file.drop(file[to_drop], axis=1)
```
